```python
a = open('./raw/ml-100k/u.data', 'rb').readlines()
```

```python
age_list = []
for i in a:
    i= i.decode('utf-8').split('\t')
    age = int(i[2])
    age_list.append(age)
```

```python
# plot the histogram of the age
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab

num_bins = 30
n, bins, patches = plt.hist(age_list, num_bins, facecolor='blue', alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Number of users')
plt.title(r'Histogram of users age')
plt.show()
```

```python
# plot the boxplot of the age
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab

plt.boxplot(age_list)
plt.xlabel('Age')
plt.ylabel('Number of users')
plt.title(r'Boxplot of users age')
plt.show()
```
